# Canny Edge Detection in OpenCV

```python
# ref : https://docs.opencv.org/4.x/da/d22/tutorial_py_canny.html, https://www.geeksforgeeks.org/python/real-time-edge-detection-using-opencv-python/
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('test.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
edges = cv.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

# Sobel Derivatives for edge detection
ref : https://docs.opencv.org/3.4/d2/d2c/tutorial_sobel_derivatives.html

```python
scale = 1
delta = 0
ddepth = cv.CV_16S

img2 = cv.imread('test.jpg', cv.IMREAD_GRAYSCALE)

# Remove noise by blurring with a Gaussian filter ( kernel size = 3 )
img2 = cv.GaussianBlur(img2,(3, 3), 0, 0, borderType=cv.BORDER_DEFAULT);

img2_x = cv.Sobel(img2, ddepth, 1, 0, ksize=3, scale=scale, delta=delta, borderType=cv.BORDER_DEFAULT)
# Gradient-Y
# grad_y = cv.Scharr(gray,ddepth,0,1)
img2_y = cv.Sobel(img2_x, ddepth, 0, 1, ksize=3, scale=scale, delta=delta, borderType=cv.BORDER_DEFAULT)
 
abs_img2_x = cv.convertScaleAbs(img2_x)
abs_img2_y = cv.convertScaleAbs(img2_y)

edge = cv.addWeighted(abs_img2_x, 0.5, abs_img2_y, 0.5, 0)
plt.subplot(121),plt.imshow(img2,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edge,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```

# Laplacian operator for edge detection
https://docs.opencv.org/3.4/d5/db5/tutorial_laplace_operator.html

```python
img2 = cv.imread('test.jpg', cv.IMREAD_GRAYSCALE)
# Declare the variables we are going to use
ddepth = cv.CV_16S
# Remove noise by blurring with a Gaussian filter ( kernel size = 3 )
img2 = cv.GaussianBlur(img2,(3, 3), 0, 0, borderType=cv.BORDER_DEFAULT);
dst = cv.Laplacian(img2, ddepth, ksize=3)
abs_dst = cv.convertScaleAbs(dst)

plt.subplot(121),plt.imshow(img2,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(abs_dst,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```
